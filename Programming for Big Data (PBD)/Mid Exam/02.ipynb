```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
temperature = np.random.randint(30,45,size=(12,30))
temperature
```

```python
temp_avg = np.round(np.mean(temperature,axis=1),2)
temp_avg
```

```python
humidity = np.random.randint(40,60,size=(12,30))
humidity
```

```python
hum_avg = np.round(np.mean(humidity,axis=1),2)
hum_avg
```

```python
rainfall = np.random.randint(0,10,size=(12,30))
rainfall
```

```python
rain_avg = np.round(np.mean(rainfall,axis=1),2)
rain_avg
```

```python
print('Average Temperature per month: ')
print(temp_avg)
print()
print('Average Humidity per month: ')
print(hum_avg)
print()
print('Average Rainfall per month: ')
print(rain_avg)
print()
```
