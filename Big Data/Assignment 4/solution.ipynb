```python
import pandas as pd
import nltk 
import re
```

```python
csv_file_path = "data.csv"
df = pd.read_csv("data.csv")
df
```

#### 1.	Remove a valid email address. (l1f20bscs0001@ucp.edu.pk)


```python
def remove_email(email):
    target_email = r'\w\d\w\d{2}\w{4}\d{4}@ucp\.edu\.pk'
    email_pat = re.compile(target_email)
    return re.sub(email_pat, '(Valid email Removed)', email)
```

```python
df['Email'] = df['Email'].apply(remove_email)
df
```

#### 2.	Remove all the URLs from a given text

```python
def remove_urls(text):
    url_pattern = re.compile(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+')
    return re.sub(url_pattern, '', text)
```

```python
df['URLs'] = df['URLs'].apply(remove_urls)
df
```

#### 3.	Check phone number in a specific format (e.g. 0423-1234567).

```python
def check_phone_number(text):
    phone_pattern = re.compile(r'0\d{3}-\d{7}')
    match = re.search(phone_pattern, text)
    if match:
        return True
    else:
        return False
```

```python
df['Has_Phone_Number'] = df['Phone'].apply(check_phone_number)
df
```

#### 4.	Find all the words that start with a capital letter in a sentence and convert them to lower case 

```python
def convert_capital_to_lower(text):
    word_pattern = re.compile(r'\b[A-Z][a-z]*\b')
    words = re.findall(word_pattern, text)
    for word in words:
        text = text.replace(word, word.lower())
    return text
```

```python
df['About'] = df['About'].apply(convert_capital_to_lower)
df
```

#### 5.	Replace all occurrences of a specific word with another word in a given text 

```python
def remove_word(text):
    return re.sub('this is', 'I am', text)
```

```python
df['About'] = df['About'].apply(remove_word)
df
```

#### 6.	Remove all the numbers from a string.

```python
def remove_numbers(text):
    number_pattern = re.compile(r'\d+')
    return re.sub(number_pattern, '', text)
```

```python
df['Address'] = df['Address'].apply(remove_numbers)
df
```
