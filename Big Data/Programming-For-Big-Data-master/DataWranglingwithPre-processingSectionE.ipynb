```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
#NaN
import numpy as np
```

```python
df.replace("?", np.nan, inplace=True)
```

<h4>Calculate the average of the column </h4>

```python
avg = df["normalized-losses"].astype("float").mean(axis=0)
```

```python
avg
```

<h4>Replace "NaN" by mean value in "normalized-losses" column</h4>

```python
df["normalized-losses"].replace(np.nan, avg, inplace=True)
```

```python
df["normalized-losses"].isnull().sum()
```

```python
df['num-of-doors'].value_counts()
```

```python
m = df['num-of-doors'].value_counts().idxmax()
```

```python
m
```

```python
df["num-of-doors"].replace(np.nan, m, inplace=True)
```

```python
df["num-of-doors"].isnull().sum()
```

```python
missing_data.sum()
```

```python
df.dropna(subset=["price"], axis=0, inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```
