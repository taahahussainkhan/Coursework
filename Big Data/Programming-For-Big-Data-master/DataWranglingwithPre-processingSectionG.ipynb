<h4>Evaluating for Missing Data</h4>

The missing values are converted to Python's default. We use Python's built-in functions to identify these missing values. There are two methods to detect missing data:
<ol>
    <li><b>.isnull()</b></li>
    <li><b>.notnull()</b></li>
</ol>
The output is a boolean value indicating whether the value that is passed into the argument is in fact missing data.

```python
missing_data = df.isnull()
```

```python
missing_data.head(15)
```

```python
missing_data.sum()
```

<h3 id="deal_missing_values">Deal with missing data</h3>
<b>How to deal with missing data?</b>

<ol>
    <li>drop data<br>
        a. drop the whole row<br>
        b. drop the whole column
    </li>
    <li>replace data<br>
        a. replace it by mean<br>
        b. replace it by frequency<br>
        c. replace it based on other functions
    </li>
</ol>

```python
df["normalized-losses"].isnull().sum()
```

```python
#df.dtypes
avg = df["normalized-losses"].astype("float").mean(axis=0)
avg
```

```python
df["normalized-losses"].replace(np.nan, avg, inplace=True)
```

```python
df.head(15)
```

```python
df["normalized-losses"].isnull().sum()
```

```python
df["num-of-doors"].value_counts()
```

```python
m = df["num-of-doors"].value_counts().idxmax()
m
```

```python
df["num-of-doors"].replace(np.nan, m, inplace=True)
```

```python
df["num-of-doors"].isnull().sum()
```

<h4>Finally, let's drop all rows that do not have price data</h4>

```python
df.shape
```

```python
df.dropna(subset=["price"], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
```
