<h4>Multiple plots in one figure</h4>

```python
#plt.subplot(row,col,CellNumber)
plt.subplot(1,2,1)
plt.plot(x,y,"ks")
plt.title("1st subplot")

#2nd Cell
plt.subplot(1,2,2)
plt.plot(x,y,"rD")
plt.title("2nd subplot")

plt.suptitle("Subplot testing")
plt.show()
```

```python
#plt.subplot(row,col,CellNumber)
plt.subplot(3,1,1)
plt.plot(x,y,"ks")
plt.title("1st subplot")

#2nd Cell
plt.subplot(3,1,2)
plt.plot(x,y,"rD")
plt.title("2nd subplot")

#3rd Cell
plt.subplot(3,1,3)
plt.plot(x,y,"r*")
plt.title("3rd subplot")

plt.suptitle("Subplot testing")
plt.show()
```

<h3>Bar Graphs</h3>

```python
sub = ["Programming","DSA","Prob","DB","DE"]
mks = [91, 89, 92, 85, 74]
rev = [5, 7, 5, 8, 10]
plt.bar(sub, mks, yerr=rev, color='cyan')
plt.title("Semester 4 Result")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
sub = ["Programming","DSA","Prob","DB","DE"]
mks = [91, 89, 92, 85, 74]
rev = [5, 7, 5, 8, 10]
plt.barh(sub,mks,xerr = rev, color='red')
plt.title("Semester 4 Result")
plt.ylabel("Subjects")
plt.xlabel("Marks")
plt.show()
#xerr is for horizontal variance and yerr is for vertical variance. 
#plt.bar -- is for vertical bar
#plt.barh -- is for horizontal bar....
```

<h2>Pie Charts in matplotlib</h2>

```python
sub = ["Programming","DSA","Prob","DB","DE"]
mks = [91, 89, 92, 85, 74]
ex = [0.1,0,0.1,0.1,0]
plt.pie(mks,labels=sub,explode=ex,shadow=True,startangle=45,autopct='%1.2f')
plt.axis('equal')
plt.legend(title='Semester 4 result')
plt.show()
```
