<h1>What is Correlation</h1>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
path = 'cleaned_data1.csv'
df = pd.read_csv(path)
df.head()
```

```python
#cov(X, Y) = (sum (x - mean(X)) * (y - mean(Y)) )/(N-1)
#X = Engine Capacity
#Y = Price
df.corr()
```

```python
#Positive Correlation
sns.regplot(x='engine-size',y='price',data=df)
```

```python
df[['engine-size','price']].corr()
```

```python
df[['highway-mpg','price']].corr()
```

```python
#Negative Correlation
sns.regplot(x='highway-mpg',y='price',data=df)
```

```python
df.corr()
```

```python
df[['compression-ratio','bore']].corr()
```

```python
#Neutral Correlation
sns.regplot(x='bore',y='compression-ratio',data=df)
```
