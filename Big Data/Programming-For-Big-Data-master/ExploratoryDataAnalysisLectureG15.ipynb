<h1>What is Correlation</h1>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
path = 'cleaned_data1.csv'
df = pd.read_csv(path)
df.head()
```

```python
#cov(X, Y) = (sum (x - mean(X)) * (y - mean(Y)) ) /(N-1)
#x = Engine-Size
#y = Price 
df.corr()
```

```python
df[['engine-size','price']].corr()
```

```python
#positive Correlation
sns.regplot(x='engine-size',y='price',data=df)
```

```python
df[['highway-mpg','horsepower']].corr()
```

```python
#Negative Correlation
sns.regplot(x='highway-mpg',y='horsepower',data=df)
```

```python
sns.regplot(x='city-mpg',y='horsepower',data=df)
```

```python
df.corr()
```

```python
df[['compression-ratio','bore']].corr()
```

```python
#Neutral Correlation....
sns.regplot(x='compression-ratio',y='bore',data=df)
```

```python
sns.boxplot(x='body-style',y='price',data=df)
```

```python
sns.boxplot(x='engine-location',y='price',data=df)
```
