<h2> Data Visulaization using Matplotlib </h2>

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1, 2, 3, 4],[1, 4, 9, 16])
plt.show()
```

```python
x = np.arange(1,5)
#x = [1,2,3,4]
y = x**3
#y = [1,8,27,64]
plt.figure(figsize=(8,5))
plt.plot([1,2,3,4],[1,4,9,16],"go",x, y,"r^")
plt.title("Plot Testing")
plt.ylabel("Y Label")
plt.xlabel("X label")
plt.show()
```

```python
plt.subplot(1,3,1)
#plt.subplot(row,col,cellNumber)
plt.plot([1,2,3,4],[1,2,9,16],"go")
plt.title("1st subplotting")

plt.subplot(1,3,2)
plt.plot([1,2,3,4],[1,4,9,16],"r^")
plt.title("2nd subplotting")

plt.subplot(1,3,3)
plt.plot([1,2,3,4],[1,4,9,16],"g^")
plt.title("3rd subplotting")

plt.suptitle("Sub Plot Testing")
plt.show()
```

```python
plt.subplot(2,1,1)
#plt.subplot(row,col,cellNumber)
plt.plot([1,2,3,4],[1,2,9,16],"go")
plt.title("1st subplotting")

plt.subplot(2,1,2)
plt.plot([1,2,3,4],[1,4,9,16],"r^")
plt.title("2nd subplotting")

plt.suptitle("Sub Plot Testing")
plt.show()
```

```python
sub = ["Programming","DSA","DB","Prob","DE"]
mks = [88, 75, 86, 92, 82]
plt.bar(sub,mks,color='cyan')
plt.title('Semester 4 Result')
plt.ylabel("Marks")
plt.xlabel("Subjects")
plt.show()
```

```python
sub = ["Programming","DSA","DB","Prob","DE"]
mks = [88, 75, 86, 92, 82]
v = [5, 4, 7, 10, 6]
plt.barh(sub,mks,xerr = v, color='cyan')
plt.title('Semester 4 Result')
plt.xlabel("Marks")
plt.ylabel("Subjects")
plt.show()
```

```python
sub = ["Programming","DSA","DB","Prob","DE"]
mks = [88, 75, 86, 92, 82]
sp = [0.1,0,0.1,0.1,0]
plt.pie(mks,explode =sp, labels=sub,shadow=True, startangle=90,autopct='%1.2f')
plt.axis('equal')
plt.legend(title='Semester 4 Result')
plt.show()
```
